# Question 11

```python
import scipy.stats as stats
import numpy as np

# Given data
sample_mean = 200  # average weight in pounds
sample_std = 30    # standard deviation of the sample
sample_size = 2000  # sample size
confidence_levels = [0.94, 0.98, 0.96]  # confidence levels

# Calculate the standard error of the mean (SEM)
sem = sample_std / np.sqrt(sample_size)

# Calculate the degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the t-values for the specified confidence levels
t_values = [stats.t.ppf((1 + confidence) / 2, degrees_of_freedom) for confidence in confidence_levels]

# Calculate the margin of error for each confidence level
margins_of_error = [t_value * sem for t_value in t_values]

# Calculate the confidence intervals
confidence_intervals = [(sample_mean - margin, sample_mean + margin) for margin in margins_of_error]

# Display the results
for confidence, interval in zip(confidence_levels, confidence_intervals):
    print(f"{confidence * 100}% Confidence Interval: {interval}")
```

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('Cars.csv')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Display summary statistics
print(df['MPG'].describe())

# Visualize the distribution
sns.histplot(df['MPG'], kde=True)
plt.title('MPG Distribution')
plt.show()
```

```python
from scipy.stats import shapiro, anderson

# Shapiro-Wilk test for normality
stat, p_value = shapiro(df['MPG'])
print(f'Shapiro-Wilk Test: Statistic={stat}, p-value={p_value}')

# Anderson-Darling test for normality
result = anderson(df['MPG'])
print(f'Anderson-Darling Test: Statistic={result.statistic}, Critical Values={result.critical_values}')
```

In conclusion, both the Shapiro-Wilk test and the Anderson-Darling test suggest that the MPG of cars in your dataset does not significantly deviate from a normal distribution.

```python
from scipy.stats import norm

# Function to calculate Z-score for a given confidence interval
def z_score(confidence_interval):
    # Convert the confidence interval to a one-sided value
    one_sided_ci = (1 + confidence_interval) / 2
    # Calculate the Z-score using the percent-point function (ppf) of the standard normal distribution
    z = norm.ppf(one_sided_ci)
    return z

# Confidence intervals
confidence_intervals = [0.90, 0.94, 0.60]

# Calculate Z-scores
z_scores = [z_score(ci) for ci in confidence_intervals]

# Display the results
for ci, z in zip(confidence_intervals, z_scores):
    print(f"{ci*100:.0f}% Confidence Interval Z-score: {z:.3f}")
```

# Question 23

```python
from scipy.stats import t
import numpy as np

# Given values
confidence_intervals = [0.95, 0.96, 0.99]
sample_size = 25

# Degrees of freedom for a t-distribution with n-1 degrees
degrees_of_freedom = sample_size - 1

# Calculate t-scores
t_scores = [t.ppf(1 - (1 - ci) / 2, df=degrees_of_freedom) for ci in confidence_intervals]

# Display the results
for ci, t_score in zip(confidence_intervals, t_scores):
    print(f"{ci*100:.0f}% Confidence Interval t-score: {t_score:.3f}")
```
